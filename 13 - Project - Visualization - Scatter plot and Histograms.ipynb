# Project - Horsepower vs Torque and Roll Dice
### Goal
- Learn how to make scatter plot and histograms

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import random
```

```python
import csv
```

```python
with open("files/cars.csv", "r") as f:
    csv_reader = csv.DictReader(f)
    cars = list(csv_reader)
```

```python
len(cars)
```

```python
cars[0]
```

```python
x = [float(car['Horsepower']) for car in cars]
y = [float(car['Torque']) for car in cars]
c = [float(car['Fuel City mpg']) for car in cars]

fig, ax = plt.subplots()
ax.scatter(x, y, c=c, alpha=.25)
ax.set_xlabel("Horsepower")
ax.set_ylabel("Torque")
```

### Roll Dice

```python
def role_dice():
    return random.randint(1, 6) + random.randint(1, 6)
```

```python
role_dice()
```

```python
data = [role_dice() for i in range(500)]
```

```python
fig, ax = plt.subplots()

ax.hist(data, bins=11)
```

```python
probability = []
for i in range(1, 7):
    for j in range(1, 7):
        probability.append(i + j)
```

```python
fig, ax = plt.subplots()

ax.hist(probability, bins=11)
```
