# Matplotlib
### Goal
- Learn to visualize data charts

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.plot([1, 2, 3, 4])
plt.ylabel('some numbers')
```

```python
fig, ax = plt.subplots()
ax.plot([1, 2, 3, 4])
ax.set_ylabel("Some numbers")
```

```python
x = [1, 2, 3, 4, 5, 6]
y = [2, 1, 2, 3, 3, 4]
c = [2, 3, 1, 2, 5, 2]
```

```python
fig, ax = plt.subplots()
ax.scatter(x, y, c=c, alpha=.75)
ax.set_xlabel("X label")
ax.set_ylabel("Y lable")
```

```python
data = [1, 2, 2, 3, 2, 1, 1, 1, 2, 3, 2, 3, 1, 1]
```

```python
fig, ax = plt.subplots()

ax.hist(data, bins=3)
```
